```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age",y="Annual Income (k$)",data=df)
plt.title("Age vs Annual Income")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.show()
```

```python
def skewness_names(skewness):
    if skewness>0:
        return "Right skewed"
    elif skewness<0:
        return "Left skewed"
    else:
        return "Normal"
    
print("Skewness of Age:",skewness_names(df['Age'].skew()))
print("Skewness of Annual Income:",skewness_names(df['Annual Income (k$)'].skew()))
print("Skewness of Spending Score:",skewness_names(df['Spending Score (1-100)'].skew()))
```
